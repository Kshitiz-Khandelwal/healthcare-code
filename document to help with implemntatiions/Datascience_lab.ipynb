# Lab 3: Information Retrieval using API

## 3.1. Retrieving articles & author information from PUBMED

```python
import requests
import json
import xml.etree.ElementTree as ElementTree
import pandas as pd
```

```python
def fetch_pubmed_articles(query, rows=50):
    url = 'https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi'
    params = {
        'db': 'pubmed',
        'term': query,
        'retmax': rows,
        'retmode': 'xml',
    }

    try:
        response = requests.get(url, params=params)
        response.raise_for_status()

        # Parse XML response
        root = ElementTree.fromstring(response.content)

        # Extract article IDs
        ids = [id_elem.text for id_elem in root.findall(".//Id")]

        if not ids:
            print("No articles found.")
            return []

        # Fetch article details using the IDs
        details_url = 'https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esummary.fcgi'
        details_params = {
            'db': 'pubmed',
            'id': ','.join(ids),
            'retmode': 'xml',
        }

        details_response = requests.get(details_url, params=details_params)
        details_response.raise_for_status()

        details_root = ElementTree.fromstring(details_response.content)
        articles = []

        for docsum in details_root.findall(".//DocSum"):
            title_elem = docsum.find("Item[@Name='Title']")
            authorlist_elem = docsum.find("Item[@Name='AuthorList']")

            title = title_elem.text if title_elem is not None else "No title"

            # AuthorList contains multiple Item elements for each author
            authors = []
            if authorlist_elem is not None:
                for author in authorlist_elem.findall("Item"):
                    if author.text:
                        authors.append(author.text)
            authors_str = ", ".join(authors) if authors else "No authors"

            articles.append({'title': title, 'authors': authors_str})

        return articles

    except requests.exceptions.RequestException as e:
        print(f"Error fetching data from PubMed: {e}")
        return None
```

```python
# Example usage - Fetching only the title information
query = "cancer research"
articles = fetch_pubmed_articles(query)

if articles:
    for article in articles:
        print(f"Title: {article['title']}")
        print(f"Authors: {article['authors']}")
        print("="*50)
```

```python
# Optional: save to CSV using pandas
df = pd.DataFrame(articles)
df.to_csv("pubmed_articles.csv", index=False)
print("Saved metadata to pubmed_articles.csv")
```

## 3.2. Retrieving articles from crossref (metadata)

```python
def fetch_crossref(query, rows=10):
    url = 'https://api.crossref.org/works'
    params = {
        'query.bibliographic': query,
        'rows': rows,
    }

    try:
        resp = requests.get(url, params=params)
        resp.raise_for_status()
        return resp.json()
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data from Crossref: {e}")
        return None

def parse_crossref(json_data):
    items = json_data.get('message', {}).get('items', [])
    parsed = []
    for entry in items:
        title = entry.get('title', [''])[0]
        authors = [f"{a.get('given', '')} {a.get('family', '')}".strip()
                   for a in entry.get('author', [])]
        doi = entry.get('DOI', '')
        url = entry.get('URL', '')
        published = entry.get('published-print', {}).get('date-parts', [[]])[0]
        published_date = "-".join(str(d) for d in published) if published else ""

        parsed.append({
            'Title': title,
            'Authors': ", ".join(authors),
            'DOI': doi,
            'URL': url,
            'Published Date': published_date
        })

    return parsed
```

```python
query = "machine learning"
rows = 40

json_data = fetch_crossref(query, rows)
if json_data:
    parsed_data = parse_crossref(json_data)

    # Save to CSV using pandas in one line:
    pd.DataFrame(parsed_data).to_csv("crossref_results.csv", index=False)

    print("Saved results to crossref_results.csv")
else:
    print("No data fetched.")
```

# Lab 4: Handling missing values

## 4.1. Check missing values (after replacing zeros with NaN)

```python
# Q1: Check missing values (after replacing zeroes with NaN)
df_missing = df_pima.copy()
df_missing[zero_cols] = df_missing[zero_cols].replace(0, np.nan)
print("\n[Lab4-Q1] Missing values per column:")
print(df_missing.isnull().sum())
```

```python
df_missing.head()
```

## 4.2. Impute Manually (mean for continous variables)

```python
# Q2: Impute manually (mean for continuous variables)
df_manual = df_missing.copy()
for c in zero_cols:
    df_manual[c].fillna(df_manual[c].mean(), inplace=True)
print("\n[Lab4-Q2] Missing values after manual imputation:")
print(df_manual.isnull().sum())
```

```python
df_manual.head()
```

## 4.3. Impute using sklearn SimpleImputer

```python
# Q3: Impute using sklearn SimpleImputer
df_sklearn = df_missing.copy()
df_sklearn[zero_cols] = SimpleImputer(strategy='median').fit_transform(df_sklearn[zero_cols])
print("\n[Lab4-Q3] Missing values after SimpleImputer:")
print(df_sklearn.isnull().sum())
```

```python
df_sklearn.head()
```

## 4.4. Clean and Save dataset

```python
# Q4: Clean and save dataset
def clean_pima(df):
    df = df.copy()
    df.columns = [col.lower().strip().replace(" ", "_") for col in df.columns]
    df = df.drop_duplicates()
    df[zero_cols] = SimpleImputer(strategy='median').fit_transform(df[zero_cols])
    df = df[(df['age'] >= 21) & (df['bmi'] <= 70)]
    return df
```

```python
clean_df = clean_pima(df_missing)
clean_df.to_csv("cleaned_pima.csv", index=False)
print("\n[Lab4-Q4] Cleaned dataset saved to 'cleaned_pima.csv'")
```
